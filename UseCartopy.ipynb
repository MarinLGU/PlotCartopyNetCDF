```python
from pylab import *
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature
import netCDF4 as nc
```

Importing and processing dataset

```python
ds=nc.Dataset('data_for_Marin/PLIO_end.nc')
thk=ds.variables['thk'][0,:]
topg=ds.variables['topg'][0,:]
x=ds.variables['x'][:]
y=ds.variables['y'][:]
X,Y=np.meshgrid(x,y)
```

Defining Projection

```python
epsg3031=ccrs.SouthPolarStereo(true_scale_latitude=-71)
```

Plotting ice-thickness

```python
fig=plt.figure()
ax1=plt.axes(projection=epsg3031)
# Limit the map to -60 degrees latitude and below.
ax1.set_extent([-180, 180, -90, -60], ccrs.PlateCarree())
ax1.coastlines(resolution='110m')
ax1.add_feature(cartopy.feature.LAND)
ax1.add_feature(cartopy.feature.OCEAN)
ax1.gridlines()
THK=plt.contourf(X,Y,thk)
plt.colorbar(THK)
plt.savefig('icethk', dpi=None, facecolor='w', edgecolor='w',
        orientation='portrait', papertype=None, format=None,
        transparent=False, bbox_inches=None, pad_inches=0.1,
        frameon=None)
plt.show()

```

Plotting Ground Level

```python
ax2=plt.axes(projection=epsg3031)
ax2.set_extent([-180, 180, -90, -60], ccrs.PlateCarree())
ax2.coastlines(resolution='110m')
ax2.add_feature(cartopy.feature.LAND)
ax2.add_feature(cartopy.feature.OCEAN)
ax2.gridlines()
TOPG=plt.contourf(X,Y,topg)
plt.colorbar(TOPG)
plt.savefig('topg')
plt.show()
```
